# QUESTÃO 5 


## Porcentagem de pessoas que sobreviveram e morreram, agrupadas por gênero (10 pts)

### RESPOSTA


```python
# Porcentagem de pessoas que sobreviveram e morreram, agrupadas por gênero (10 pts)


#  Imprime uma mensagem para indicar que será exibido um dataframe com informações sobre a porcentagem de pessoas que 
#  sobreviveram e morreram por gênero.
print("\nDATAFRAME DA PORCENTAGEM DE PESSOAS QUE SOBREVIVERAM E MORRERAM POR GÊNERO\n")

#  Cria um objeto DataFrame que agrupa os dados do DataFrame 
#  "df_titanic" por "Survived" e "Sex", calcula a porcentagem de pessoas
#  que sobreviveram e morreram por gênero e, em seguida, multiplica por 100 para obter um valor percentual.

sobreviventes = df_titanic.groupby('Survived')['Sex'].value_counts(normalize=True) * 100

#  Renomeia a coluna "value" para "%" e redefine os índices do objeto 
#  DataFrame "sobreviventes" para ser uma coluna no DataFrame "resultado".

resultado = sobreviventes.rename('%').reset_index()

# Arredonda o valor da coluna "%" para duas casas decimais.
resultado['%'] = resultado['%'].round(2)

# Renomeia as colunas "Survived" e "Sex" para "Sobreviveram" e "Sexo", respectivamente, no objeto DataFrame "resultado".
resultado = resultado.rename(columns={'Survived': 'Sobreviveram', 'Sex': 'Sexo'})

# Substitui os valores "0" e "1" na coluna "Sobreviveram" por "NÃO" e "SIM", respectivamente, 
# e substitui os valores "male" e "female" na coluna "Sexo" por "masculino" e "feminino", respectivamente.
resultado.replace({'Sobreviveram': {0: 'NÃO', 1: 'SIM'},'Sexo': {'male': 'masculino', 'female': 'feminino'}}, inplace=True)

# Exibe o objeto DataFrame "resultado" na saída do programa.
display(resultado)


```

## Porcentagem de pessoas que sobreviveram em cada classe, agrupadas por local de embarque

### RESPOSTA

```python
## Porcentagem de pessoas que sobreviveram em cada classe, agrupadas por local de embarque

print("\nDATAFRAME DA PORCENTAGEM DE PESSOAS QUE SOBREVIVERAM EM CADA CLASSE POR LOCAL DE EMBARQUE \n")

# Filtra o DataFrame para obter as linhas de sobreviventes e calcula a porcentagem de sobreviventes por classe e local de embarque.
sobreviventes_classe = df_titanic[df_titanic['Survived'] == 1].groupby('Embarked')['Pclass'].value_counts(normalize=True) * 100

# Renomeia as colunas e reorganiza o DataFrame para apresentar a porcentagem de sobreviventes por classe e local de embarque.
resultado = sobreviventes_classe.rename('%').reset_index()
resultado['%'] = resultado['%'].round(2)
resultado = resultado.rename(columns={'Pclass': 'Classe', 'Embarked': 'Local de Embarque'})

# Exibe o DataFrame "resultado".
display(resultado)

```

## Remoção de passageiros com idade desconhecida

### RESPOSTA

```python
## Remoção de passageiros com idade desconhecida

print("\n NOVO DATAFRAME COM REMOÇÃO DE PASSAGEIRSOS COM IDADE DESCONHECIDA \n")

# Cria um novo DataFrame chamado "novo_titanic" a partir do DataFrame "df_titanic".
# Remove todas as linhas que possuem valores faltantes (NaN) na coluna "Age".
novo_titanic = df_titanic.dropna(subset=['Age'])

# Exibe o novo DataFrame "novo_titanic".
display(novo_titanic)




```

## Média de idade das mulheres em cada classe, agrupadas por local de embarque

### RESPOSTA

```python
## Média de idade das mulheres em cada classe, agrupadas por local de embarque

print("\nDATAFRAME COM A MEDIA DA IDADE DAS MULHERES EM CADA CLASSE AGRUPADO POR PORTÃO DE EMBARQUE \n")

# Cria um novo DataFrame "sobreviventes" agrupando os dados por gênero e sobrevivência.
# Calcula a porcentagem de sobreviventes e de mortos para cada gênero.
sobreviventes = df_titanic.groupby('Survived')['Sex'].value_counts(normalize=True) * 100

# Renomeia a coluna com as porcentagens para facilitar a leitura.
# Arredonda os valores das porcentagens para duas casas decimais.
# Renomeia as colunas "Survived" e "Sex" para "Sobreviveram" e "Sexo", respectivamente.
# Substitui os valores numéricos da coluna "Sobreviveram" por "SIM" e "NÃO".
# Substitui os valores "male" e "female" da coluna "Sexo" por "masculino" e "feminino".
resultado = sobreviventes.rename('%').reset_index()
resultado['%'] = resultado['%'].round(2)
resultado = resultado.rename(columns={'Survived': 'Sobreviveram', 'Sex': 'Sexo'})
resultado.replace({'Sobreviveram': {0: 'NÃO', 1: 'SIM'},'Sexo': {'male': 'masculino', 'female': 'feminino'}}, inplace=True)

# Exibe o DataFrame "resultado".
display(resultado)
```

# QUESTÃO 9


## Adição de coluna booleana para identificar menores de idade

### RESPOSTA

```python
## Adição de coluna booleana para identificar menores de idade

print("\n COPIA DO DATAFRAME TITANIC COM ADIÇAO DE UMA NOVA COLUNA VERIFICANDO SE É MENOR DE IDADE \n")

#Faz uma copia do df_titanic
novo_titanic = df_titanic.copy()

#cria uma nova coluna (menor_idade)  que verifica se é menor que 18 anos
novo_titanic.loc[:, 'menor_idade'] = novo_titanic['Age'] < 18

display(novo_titanic)

```

# QUESTÃO 10

##Porcentagem de menores e adultos que sobreviveram em cada classe, agrupadas por local de embarque

### RESPOSTA

```python
print("\n DATAFRAME COM A PORCENTAGEM DE MENORES E ADULTOS QUE SOBREVIVERAM EM CADA CLASSE, AGRUPADA POR LOCAL DE EMBARQUE \n")

# Cria um novo DataFrame "sobrevivente_idade" com as colunas "Embarked", "Pclass", "Survived" e "menor_idade".

sobrevivente_idade = novo_titanic[['Embarked', 'Pclass', 'Survived', 'menor_idade']]

# Agrupa os dados por "Embarked", "Pclass" e "menor_idade", e calcula a soma de "Survived".
# Calcula a porcentagem de sobreviventes em relação ao total de passageiros para cada grupo.
# Arredonda os valores das porcentagens e renomeia a coluna com os valores somados de "Survived" para facilitar a leitura.

grupo = sobrevivente_idade.groupby(['Embarked', 'Pclass', 'menor_idade'], group_keys=False)['Survived'].sum()
resultado = grupo.groupby(level=[0, 1], group_keys=False).apply(lambda x: 100 * x / float(x.sum()))

resultado = grupo.rename('%').reset_index()



# Renomeia as colunas "Pclass" e "Embarked" para "Classe" e "Local de Embarque", respectivamente.
# Substitui os valores "True" e "False" da coluna "menor_idade" por "SIM" e "NÃO", respectivamente.

resultado = resultado.rename(columns={'Pclass': 'Classe', 'Embarked': 'Local de Embarque'})
resultado.replace({'menor_idade': {False: 'NÃO', True: 'SIM'}}, inplace=True)

# Exibe o DataFrame "resultado".
display(resultado.style.format({'%': '{:.2f}%'}))
```
